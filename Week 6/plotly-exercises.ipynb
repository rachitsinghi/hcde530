# Plotly Walkthrough

Plotly is an open source graphing library for Python. This module shows basic visualizations that can be done using Plotly. 

>_Built on top of the Plotly JavaScript library (plotly.js), plotly enables Python users to create beautiful interactive web-based visualizations that can be displayed in Jupyter notebooks, saved to standalone HTML files, or served as part of pure Python-built web applications using Dash. The plotly Python library is sometimes referred to as "plotly.py" to differentiate it from the JavaScript library._


```python
import plotly.express as px # plotly library
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

Plotly offers a variety of [visualizations](https://plotly.com/python/). From basic graphs to advanced interactive visualizations. We will go through different type of visualizations plotly is capable of. 

## Bar graph

With px.bar, each row of the DataFrame is represented as a rectangular mark. Customization documentation can be found [here](https://plotly.com/python/styling-plotly-express/). 

Every visualization created with plotly offers interactivy right away. If you hover over the bars a tooltip with information appears. Also, on the top right corner there are multiple option to zoom, pan, select, and more! You can even download a png version of the visualization. Try it yourself!

```python
px.bar(y=[1, 1, 2, 3, 5, 8, 13, 21], title = "Example Bar graph", labels = {"x":"X axis", "y":"Y axis"})
```

You can also create a figure object to have more flexibility later on. 

```python
fig = px.bar(y=[1, 1, 2, 3, 5, 8, 13, 21], title = "Example Bar graph", labels = {"x":"X axis", "y":"Y axis"})
fig.show()
```

## Scatter Plot

With px.scatter, each data point is represented as a marker point, whose location is given by the x and y columns. Here we are using the [Iris Dataset](https://archive.ics.uci.edu/ml/datasets/iris) to plot sepal width vs length.

```python
df = px.data.iris()
fig = px.scatter(df, 
                 x="sepal_width", 
                 y="sepal_length", 
                 color="species",
                 size='petal_length', 
                 hover_data=['petal_width'], # We can add more data to the tooltip
                 labels={"sepal_length": "Sepal length", 
                         "sepal_width": "Sepal width",
                         "species": "Species"}, 
                 title ="Sepal width vs Sepal length")
fig.show()
```

## Line graphs

With px.line, each data point is represented as a vertex (which location is given by the x and y columns) of a polyline mark in 2D space. Here we are showing the life expectancy of every country in America.

```python
df = px.data.gapminder().query("continent=='Americas'")
fig = px.line(df, x="year", y="lifeExp", color='country', title="Life expectancy in America countries")
fig.show()
```

## Histograms

In plotly, a histogram is an aggregated bar chart, with several possible aggregation functions (e.g. sum, average, count...). Also, the data to be binned can be numerical data but also categorical or date data. By default, the number of bins is chosen so that this number is comparable to the typical number of samples in a bin. This number can be customized, as well as the range of values.

```python
x = np.random.randn(500)
fig = px.histogram(x=x, histnorm='probability')

fig.show()
```

## 3D surface Plot

There are a wide variety of visualizations possible with plotly. We will be using mostly plotly express. `plotly.graph_objects`contains more complex visualizations such as this cool 3D surface plot of Mt. Bruno. You can interact with it to see it from all angles.

```python
import plotly.graph_objects as go

z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

# UFO Sightings dataset

Let's create some visualizations using plotly with a dataset of [UFO sigthings](https://www.kaggle.com/thaddeussegura/ufo-sightings) in the US.

```python
df = pd.read_csv("clean_ufo_data.csv")
df.tail()
```

The dates seem to have different formats. Let's change them all to datetime and put any unrecognized format as "Not a Time" (NaT) using the `errors='coerce'` option. This can take some time to finish.

```python
df["occurred_date_time"] = pd.to_datetime(df["occurred_date_time"], errors='coerce')
```

```python
df.tail()
```

They seem to have datetime format now, but a lot of them are NaT. We will drop any rows where date time is NaT.

```python
df = df.dropna(subset=['occurred_date_time'])
df.tail()
```

Now it looks much better!

Let's group the data by year and state. Using `groupby` we can count how many ocurrences there are for any combination of year an state.

```python
df_grouped = df.groupby([df["occurred_date_time"].dt.year, df["state"]])["id"].count().reset_index(name="sightings")
df_grouped.rename(columns={'occurred_date_time':'year'}, inplace=True) # we will rename this column to be more representative (this is not necessary)
df_grouped.head()
```

Now we have a dataframe including year, state and number of sightings. Let's take a look into the data for Washington state.

```python
df_wa = df_grouped[df_grouped["state"]=="WA"]
df_wa.tail(5)
```

Let's create a bar chart of the sightings per year for Washington state

```python
fig = px.bar(df_wa,x="year",y="sightings", title="UFO sightings in Washington state")
fig.show()
```

Creating a scatter plot of the same data is easy!

```python
fig = px.scatter(df_wa, x="year", y="sightings", title="UFO sightings in Washington state")
fig.show()
```

We want to see this data for all of the states at the same time. Let's create a scatter plot that includes all of the states.

```python
fig = px.scatter(df_grouped, x="year", y="sightings", color="state", title="UFO sightings in USA per state")
fig.show()
```

We can also create a new column that includes the cumulative sum of all the sightings.

```python
df_grouped['total'] = df_grouped.groupby(['state'])['sightings'].apply(lambda x: x.cumsum())
df_grouped.tail()
```

Using this data we can plot total UFO sightings per state using a line graph. You can filter data in the visualization using the state key.

```python
fig = px.line(df_grouped, x="year", y="total", color='state', title = "Total UFO Sightings per state")
fig.show()
```

There's too much occlusion. Let's do the same visualization but with the top 5 states with most sightings.

```python
top5 = df_grouped.groupby("state")["sightings"].sum().sort_values(ascending=False).head(5).reset_index()["state"]
top5
```

```python
fig = px.line(df_grouped[df_grouped["state"].isin(top5)], 
              x="year", 
              y="total", 
              color='state', 
              title = "Top 5 states with most UFO sightings")
fig.show()
```

There is still some occlusion. If we have the data per state maybe using a map is better to show all the information at the same time.

## Choropleth map

Using plotly we will create a choropleth map that show sightings per state. Let's show all the sightings in the year 2005. We will be using a color scale. You can read more about plotly built-in color scales [here](https://plotly.com/python/builtin-colorscales/).

```python
year_df = df_grouped[df_grouped["year"] == 2005]

fig = px.choropleth(year_df, 
                    locations = "state",  
                    locationmode="USA-states", 
                    color='sightings',      
                    color_continuous_scale="bluyl",
                    labels={'sightings':'UFO Sightings', "state":"State"},
                    scope="usa")

fig.update_layout(title= "UFO Sightings in 2005", 
                  margin={"r":20,"t":100,"l":20,"b":20})

fig.show()
```

Let's also show the year. We can do that by adding a slider using the `animation_frame` option.

```python
fig = px.choropleth(df_grouped, 
                    locations=df_grouped['state'],  
                    locationmode="USA-states", 
                    color='sightings',      
                    color_continuous_scale="bluyl",
                    scope="usa",
                    animation_frame="year", 
                    title= "UFO Sightings per year",
                    labels={'sightings':'UFO Sightings', "state":"State"})
fig.show()
```

Now we can move through the years! You can see that the range of UFO sightings changes for each year. Comparing between years is more difficult that way. To fix this we can add a fixed range for all of the years with the option `range_color`.

```python
fig = px.choropleth(df_grouped, 
                    locations=df_grouped['state'],  
                    locationmode="USA-states", 
                    color='sightings',      
                    color_continuous_scale="bluyl",
                    range_color=(0, df_grouped["sightings"].max()), # Fixed color range
                    scope="usa",
                    animation_frame="year", 
                    title= "UFO Sightings per year",
                    labels={'sightings':'UFO Sightings', "state":"State"})
fig.update_layout(margin={"r":20,"t":100,"l":20,"b":20})
fig.show()
```

Now we have an interactive visualization that shows all the sightings per year and state at the same time.

Create a new (fun) additional visualization of a type that we did not cover in class using the same dataset. You can look for inspirations [here](https://plotly.com/python/)! Explain your chart, including choice of dimensions, values, and encodings in comments or a markdown text box.

```python
# Exercise 4: 3D Bird Strike SKYSCRAPER CITY! 🏙️🐦
#
# Real 3D buildings (boxes) where each building represents a combination of
# Flight Phase × Visibility, and the building's HEIGHT shows how many bird
# strikes happened in that combo. The taller the building, the more dangerous!
#
# Dimensions:
#   - X axis: Flight Phase (each phase = a row of buildings)
#   - Y axis: Visibility (Day, Night, Dawn, Dusk)
#   - Z axis (height): Number of incidents
# Color: Aircraft Damage (red glowing = damaged, green = no damage)
#
# It's a futuristic data city you can fly around in! 🌃✨

import pandas as pd
import plotly.graph_objects as go
import numpy as np

# Load and clean
bird_df = pd.read_csv("bird_strikes.csv")
bird_df["Damage Status"] = bird_df["Aircraft Damage"].map({1: "Damaged", 0: "No Damage"})
df = bird_df.dropna(subset=["Flight Phase", "Visibility", "Damage Status"])

grouped = df.groupby(
    ["Flight Phase", "Visibility", "Damage Status"]
)["Record ID"].count().reset_index(name="Incidents")

phases = list(grouped["Flight Phase"].unique())
visibilities = ["DAY", "DUSK", "DAWN", "NIGHT"]
phase_to_x = {p: i * 2 for i, p in enumerate(phases)}  # spaced apart
vis_to_y = {v: i * 2 for i, v in enumerate(visibilities)}

damage_colors = {"Damaged": "#FF3B3B", "No Damage": "#4ECDC4"}

# Helper function: build a box (skyscraper) at position (cx, cy) with given height
def make_building(cx, cy, height, width=0.7, color="#4ECDC4", name="", hover=""):
    # 8 corners of a box
    x = [cx-width, cx+width, cx+width, cx-width, cx-width, cx+width, cx+width, cx-width]
    y = [cy-width, cy-width, cy+width, cy+width, cy-width, cy-width, cy+width, cy+width]
    z = [0, 0, 0, 0, height, height, height, height]

    # 12 triangles to form the 6 faces of the box
    i = [0, 0, 0, 0, 4, 4, 1, 1, 2, 2, 3, 3]
    j = [1, 2, 4, 3, 5, 6, 5, 6, 6, 7, 7, 4]
    k = [2, 3, 5, 4, 6, 7, 6, 2, 7, 3, 4, 0]

    return go.Mesh3d(
        x=x, y=y, z=z,
        i=i, j=j, k=k,
        color=color,
        opacity=0.92,
        flatshading=True,
        name=name,
        hovertext=hover,
        hoverinfo="text",
        showlegend=False,
        lighting=dict(
            ambient=0.4,
            diffuse=0.9,
            specular=0.6,
            roughness=0.3,
            fresnel=0.2
        ),
        lightposition=dict(x=200, y=200, z=400)
    )

# Helper: add glowing "window dots" on the building sides for nighttime city vibes ✨
def make_windows(cx, cy, height, width=0.7):
    if height < 50:
        return None  # too small for windows
    # Random window dots scattered on the front faces
    n_windows = int(min(height / 30, 60))
    np.random.seed(int(cx * 100 + cy * 10))  # consistent windows per building
    window_x, window_y, window_z = [], [], []
    for _ in range(n_windows):
        side = np.random.choice([0, 1, 2, 3])
        h = np.random.uniform(20, height - 10)
        offset = np.random.uniform(-width * 0.7, width * 0.7)
        if side == 0:    # front
            window_x.append(cx + offset); window_y.append(cy - width - 0.01); window_z.append(h)
        elif side == 1:  # back
            window_x.append(cx + offset); window_y.append(cy + width + 0.01); window_z.append(h)
        elif side == 2:  # left
            window_x.append(cx - width - 0.01); window_y.append(cy + offset); window_z.append(h)
        else:            # right
            window_x.append(cx + width + 0.01); window_y.append(cy + offset); window_z.append(h)
    return go.Scatter3d(
        x=window_x, y=window_y, z=window_z,
        mode="markers",
        marker=dict(size=2.5, color="#FFEB3B", opacity=0.9, symbol="square"),
        showlegend=False,
        hoverinfo="skip"
    )

fig = go.Figure()

# Build all the skyscrapers!
for _, row in grouped.iterrows():
    if row["Visibility"] not in vis_to_y:
        continue
    cx = phase_to_x[row["Flight Phase"]]
    cy = vis_to_y[row["Visibility"]]
    height = row["Incidents"]
    color = damage_colors[row["Damage Status"]]

    # Damaged buildings stack on top of no-damage ones (no overlap)
    if row["Damage Status"] == "Damaged":
        cy += 0.85  # offset slightly so damaged sits next to no-damage

    hover = (f"<b>{row['Flight Phase']}</b><br>"
             f"Visibility: {row['Visibility']}<br>"
             f"Status: {row['Damage Status']}<br>"
             f"Incidents: {row['Incidents']:,}")

    fig.add_trace(make_building(cx, cy, height, width=0.4, color=color, hover=hover))

    # Add windows for tall buildings
    windows = make_windows(cx, cy, height, width=0.4)
    if windows:
        fig.add_trace(windows)

# Add a "ground" plane for the city to sit on
ground_x = [-2, max(phase_to_x.values()) + 2, max(phase_to_x.values()) + 2, -2]
ground_y = [-2, -2, max(vis_to_y.values()) + 3, max(vis_to_y.values()) + 3]
ground_z = [0, 0, 0, 0]
fig.add_trace(go.Mesh3d(
    x=ground_x, y=ground_y, z=ground_z,
    i=[0, 0], j=[1, 2], k=[2, 3],
    color="#1a1a2e",
    opacity=0.8,
    showlegend=False,
    hoverinfo="skip",
    lighting=dict(ambient=0.5)
))

# Fake legend entries
for status, color in damage_colors.items():
    label = "💥 Damaged" if status == "Damaged" else "✅ No Damage"
    fig.add_trace(go.Scatter3d(
        x=[None], y=[None], z=[None],
        mode="markers",
        marker=dict(size=14, color=color, symbol="square"),
        name=label,
        showlegend=True
    ))

fig.update_layout(
    title="🏙️ Bird Strike City: A Skyscraper Skyline of Aviation Danger 🐦<br><sub>Building height = # of incidents • 🪟 lit windows = busy buildings • 🖱️ Drag to fly around</sub>",
    width=1000,
    height=780,
    template="plotly_dark",
    scene=dict(
        xaxis=dict(
            title="✈️ Flight Phase",
            tickvals=list(phase_to_x.values()),
            ticktext=list(phase_to_x.keys()),
            gridcolor="rgba(255,255,255,0.05)",
            backgroundcolor="rgb(8, 12, 30)"
        ),
        yaxis=dict(
            title="👁️ Visibility",
            tickvals=list(vis_to_y.values()),
            ticktext=list(vis_to_y.keys()),
            gridcolor="rgba(255,255,255,0.05)",
            backgroundcolor="rgb(8, 12, 30)"
        ),
        zaxis=dict(
            title="📊 # of Incidents",
            gridcolor="rgba(255,255,255,0.05)",
            backgroundcolor="rgb(8, 12, 30)"
        ),
        bgcolor="rgb(3, 5, 18)",
        camera=dict(eye=dict(x=1.8, y=1.8, z=0.7)),
        aspectratio=dict(x=2, y=1.2, z=1.5)
    ),
    margin=dict(t=100, l=0, r=0, b=20),
    legend=dict(bgcolor="rgba(0,0,0,0.6)", bordercolor="white", borderwidth=1, x=0.85, y=0.95),
    updatemenus=[
        dict(
            type="buttons",
            direction="right",
            x=0.5, y=-0.05, xanchor="center",
            buttons=[
                dict(label="🔍 Zoom In",
                     method="relayout",
                     args=[{"scene.camera.eye": dict(x=1.0, y=1.0, z=0.5)}]),
                dict(label="🔭 Zoom Out",
                     method="relayout",
                     args=[{"scene.camera.eye": dict(x=3.0, y=3.0, z=1.5)}]),
                dict(label="🏠 Reset",
                     method="relayout",
                     args=[{"scene.camera.eye": dict(x=1.8, y=1.8, z=0.7)}]),
                dict(label="🚁 Helicopter",
                     method="relayout",
                     args=[{"scene.camera.eye": dict(x=0, y=0, z=2.5)}]),
                dict(label="🚶 Street View",
                     method="relayout",
                     args=[{"scene.camera.eye": dict(x=2.8, y=0.2, z=0.15)}]),
            ]
        )
    ]
)

fig.show()
```

```python
# Exercise 4: 3D ANIMATED Flapping Bird Strikes! 🐦💫
#
# A 3D scatter plot where every data point is a bird emoji that ACTUALLY
# FLAPS ITS WINGS! We use Plotly animation frames to cycle through different
# bird emojis (wings up → wings down → wings up...) creating a flapping effect.
#
# Press the ▶ Play button to make the whole flock take flight!
#
# Dimensions:
#   - X axis: Speed (knots)
#   - Y axis: Height (altitude)
#   - Z axis: Year
# Color: Flight Phase

import pandas as pd
import plotly.graph_objects as go

# Load and clean
bird_df = pd.read_csv("bird_strikes.csv")
plot_df = bird_df.dropna(
    subset=["Speed", "Height", "Incident Year", "Flight Phase",
            "Operator", "Species Name"]
).copy()
plot_df = plot_df.sample(n=1000, random_state=42)  # smaller for smoother animation

# Two bird poses: wings up vs wings down (the flap cycle!)
# Each phase gets a pair: [wings up, wings down]
bird_flap_pairs = {
    "APPROACH":     ("🦅", "🪶"),
    "TAKEOFF RUN":  ("🐦", "🪺"),
    "LANDING ROLL": ("🦆", "🦢"),
    "CLIMB":        ("🦉", "🌙"),
    "DESCENT":      ("🦜", "🌈"),
    "EN ROUTE":     ("🕊️", "☁️"),
    "LANDING":      ("🐧", "❄️"),
    "TAXI":         ("🦤", "🪶"),
    "DEPARTURE":    ("🐤", "🐣"),
    "LOCAL":        ("🐥", "🪺"),
    "ARRIVAL":      ("🦢", "🪽"),
    "PARKED":       ("🐔", "🥚"),
}

colors = ["#FF6B6B", "#4ECDC4", "#FFD93D", "#95E1D3", "#C06EFA",
          "#FF9F1C", "#A8E6CF", "#FF8B94", "#FFAAA5", "#B5EAD7",
          "#C7CEEA", "#FFDAC1"]

# Helper: build the traces (one per flight phase) for a given wing position
def build_traces(wing_position):  # 0 = wings up, 1 = wings down
    traces = []
    for i, (phase, emoji_pair) in enumerate(bird_flap_pairs.items()):
        phase_data = plot_df[plot_df["Flight Phase"] == phase]
        if len(phase_data) == 0:
            continue
        emoji = emoji_pair[wing_position]
        traces.append(go.Scatter3d(
            x=phase_data["Speed"],
            y=phase_data["Height"],
            z=phase_data["Incident Year"],
            mode="text",
            text=[emoji] * len(phase_data),
            textfont=dict(size=20, color=colors[i % len(colors)]),
            name=f"{emoji_pair[0]} {phase}",
            hovertext=[
                f"<b>{phase}</b><br>"
                f"Airline: {op}<br>"
                f"Species: {sp}<br>"
                f"Speed: {sp_val} kts<br>"
                f"Height: {h} ft<br>"
                f"Year: {yr}"
                for op, sp, sp_val, h, yr in zip(
                    phase_data["Operator"],
                    phase_data["Species Name"],
                    phase_data["Speed"],
                    phase_data["Height"],
                    phase_data["Incident Year"]
                )
            ],
            hoverinfo="text"
        ))
    return traces

# Build the figure with initial frame (wings up)
fig = go.Figure(
    data=build_traces(0),
    frames=[
        go.Frame(data=build_traces(0), name="up"),
        go.Frame(data=build_traces(1), name="down"),
    ]
)

# Add play/pause buttons + flap controls
fig.update_layout(
    title="🐦 FLAPPING Flock of Bird Strikes ✨<br><sub>Press ▶ Play to make them flap! • 🖱️ Drag to rotate • Scroll to zoom</sub>",
    width=950,
    height=750,
    template="plotly_dark",
    scene=dict(
        xaxis_title="✈️ Aircraft Speed (knots)",
        yaxis_title="☁️ Altitude (feet)",
        zaxis_title="📅 Year",
        bgcolor="rgb(15, 20, 50)",
        xaxis=dict(gridcolor="rgba(255,255,255,0.1)"),
        yaxis=dict(gridcolor="rgba(255,255,255,0.1)"),
        zaxis=dict(gridcolor="rgba(255,255,255,0.1)"),
        camera=dict(eye=dict(x=1.2, y=1.2, z=0.8))
    ),
    margin=dict(t=100, l=0, r=0, b=80),
    legend=dict(
        bgcolor="rgba(0,0,0,0.5)",
        bordercolor="white",
        borderwidth=1
    ),
    updatemenus=[
        # Flap controls 🪽
        dict(
            type="buttons",
            direction="right",
            x=0.5,
            y=-0.05,
            xanchor="center",
            showactive=False,
            buttons=[
                dict(
                    label="▶ FLAP!",
                    method="animate",
                    args=[None, {
                        "frame": {"duration": 250, "redraw": True},
                        "fromcurrent": True,
                        "mode": "immediate",
                        "transition": {"duration": 0}
                    }]
                ),
                dict(
                    label="⏸ Land",
                    method="animate",
                    args=[[None], {
                        "frame": {"duration": 0, "redraw": False},
                        "mode": "immediate",
                        "transition": {"duration": 0}
                    }]
                ),
            ]
        ),
    ]
)

fig.show()
```
