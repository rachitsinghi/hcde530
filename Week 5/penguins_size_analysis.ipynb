## Load the CSV

Missing values in the file are the literal string **`NA`**. Passing `na_values=["NA"]` converts them to pandas missing values so numeric columns parse correctly.

```python
data_path = Path("penguins_size 2.csv")
if not data_path.is_file():
    data_path = Path("Week 5") / "penguins_size 2.csv"

if not data_path.is_file():
    raise FileNotFoundError(
        f"Could not find the CSV (tried current dir and Week 5/). CWD is {Path.cwd()}"
    )

df = pd.read_csv(data_path, na_values=["NA"])
print("Loaded from:", data_path.resolve())
```

## Question 2 — Distribution of **`flipper_length_mm`**

Flipper length is numeric (millimeters). **`value_counts()`** lists each observed flipper length and how many rows have that value (like a frequency table). Sorting by index makes the table easier to scan.

```python
df["flipper_length_mm"].value_counts(dropna=False).sort_index()
```

## Question 3 — Meaningful subset: **Adelie** penguins, **female** only

Boolean filters pick rows where **`species`** is Adelie and **`sex`** is FEMALE (same idea as filtering ratings below a threshold, but with equality on categories).

```python
adelie_female = df[(df["species"] == "Adelie") & (df["sex"] == "FEMALE")]
adelie_female.shape
```

```python
adelie_female.head(10)
```

## Question 4 — **`groupby`**: mean **`flipper_length_mm`** by **`island`**

Average flipper length for each island (all species in the full frame; missing flipper lengths are skipped automatically for the mean).

```python
df.groupby("island", observed=True)["flipper_length_mm"].mean().sort_values(ascending=False)
```

## Question 5 — Missing values

**`isnull().sum()`** counts missing cells per column.

```python
df.isnull().sum()
```

**Interpretation:** Two rows are missing all four measurement columns (and often **`sex`**) together—likely one penguin with no recorded measurements. **`sex`** has more missing values than the physical measures, so some rows have culmen/flipper/body mass but no sex label. **`species`** and **`island`** have no missing values, so every row still has species and island. Check **`sex`** with **`value_counts(dropna=False)`** if you need to spot odd labels (for example a stray `"."` in this file).
